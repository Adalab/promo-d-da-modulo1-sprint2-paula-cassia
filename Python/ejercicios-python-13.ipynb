```python
# Query 1: el primer pedido donde se ha gastado mas

mydb = mysql.connector.connect(
user='root', 
password='AlumnaAdalab',
host='127.0.0.1',
database='northwind')

cursor = mydb.cursor()

query1 = "SELECT order_id, (unit_price*quantity) AS Cantidad_gastada FROM order_details ORDER BY (unit_price*quantity) DESC" 

cursor.execute(query1)

resultado = cursor.fetchone()
resultado
```

```python
# Query 2: todos los pedidos con la cantidad gastada 

mydb = mysql.connector.connect(
user='root', 
password='AlumnaAdalab',
host='127.0.0.1',
database='northwind')

cursor = mydb.cursor()

query2 = "SELECT order_id, (unit_price*quantity) AS Cantidad_gastada FROM order_details ORDER BY (unit_price*quantity) DESC" 

cursor.execute(query2)

resultado2 = cursor.fetchall()
resultado2
```

```python
import pandas as pd
```

```python
# Query 3: numero de pedidos por dia en un dataframe
mydb = mysql.connector.connect(
user='root', 
password='AlumnaAdalab',
host='127.0.0.1',
database='northwind')

cursor = mydb.cursor()

query3 = "SELECT order_id, order_date, COUNT(order_id) FROM orders GROUP BY order_id, order_date" 

cursor.execute(query3)

resultado3 = cursor.fetchall()

df = pd.DataFrame(resultado3, columns = ['Order ID', 'Order Date', 'No. of Orders'])

df.head(20)
```

```python
# Query 4: pedidos hechos por Nancy Davolio

mydb = mysql.connector.connect(
user='root', 
password='AlumnaAdalab',
host='127.0.0.1',
database='northwind')

cursor = mydb.cursor()

query4 = 'SELECT * FROM orders WHERE employee_id IN (SELECT employee_id FROM employees WHERE first_name = "Nancy" AND last_name = "Davolio")'

cursor.execute(query4)

resultado4 = cursor.fetchall()
print(resultado4)

cursor.close()
mydb.close()
```
